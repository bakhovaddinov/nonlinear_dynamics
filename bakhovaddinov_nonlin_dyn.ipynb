```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def print_func(mu):
    X, Y = [], []
    x = 0.5
    for x in np.linspace(0, 1, 1000):
        y = my_func(x, mu)
        X.append(x)
        Y.append(y)
    plt.figure(figsize=(6, 3))
    plt.title(f"mu = {mu}")
    plt.plot(X, Y, "k", markersize = 0.05)
    plt.plot(X, X, "b", markersize = 0.05)
    plt.xlabel("x_n")
    plt.ylabel("x_n+1")
```
